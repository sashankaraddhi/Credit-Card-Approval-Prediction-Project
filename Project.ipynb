```python
print(df['Owned_Work_Phone'].value_counts())
print(df['Owned_Phone'].value_counts())
print(df['Owned_Car'].value_counts())
print(df['Owned_Realty'].value_counts())
```

```python
columns_to_drop = ['Applicant_ID','Owned_Mobile_Phone']

for col in columns_to_drop:
    if col in df.columns:
        df.drop(col, axis=1, inplace=True)
        print(f"Column '{col}' dropped successfully.")
    else:
        print(f"Column '{col}' not found in DataFrame; skipping drop operation.")

print("Current DataFrame shape:", df.shape)

print(df.info())
```
